```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import argparse
import pickle
from constants import *
# Set the font globally
# plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.family'] = 'Times New Roman'

def get_test_mrr(file_path):
    try:
        with open(file_path, 'r') as f:
            mrr = None
            for lines in f:
                if lines.startswith('\ttest AP'):
                    mrr = float(lines.strip('\n').split(':')[-1])
                    return mrr
    except FileNotFoundError:
        import pdb; pdb.set_trace()
        pass

def get_best_epoch(file_path):
    try:
        with open(file_path, 'r') as f:
            epoch = None
            for lines in f:
                if lines.startswith('Loading'):
                    epoch = int(lines.split(' ')[4])
                    return epoch
    except FileNotFoundError:
        import pdb; pdb.set_trace()
        pass

show_datasets = {
    'WIKI': 'Wikipedia',
    'REDDIT': 'REDDIT',
    'Flights': 'Flights',
    'LASTFM': 'LASTFM',
    'mooc': 'MOOC',
    'uci': 'UCI',
    'CollegeMsg': 'CollegeMsg'
}
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,10))
datasets = ['WIKI', 'REDDIT', ]
handles = []
i = 0
ax = axes[0]
for k, dataset in enumerate(datasets):
    df_all = pd.DataFrame()
    ax = axes[0][k]
    scans = ['10', '20', '50', '100']
    if dataset == 'mooc' or dataset == 'uci':
        title_str = f'{dataset.upper()}'
    else:
        title_str = f'{show_datasets[dataset]}'
    ax.set_title(title_str, x=0.5, y=1.05)
    for memory in memorys:
        sampling = 're'
        means = []
        stds = []
        results = []
        results2 = []
        for scan in scans:
            results.append(np.mean(np.array(all_data[dataset][scan]['TGAT'][sampling][memory]) ))
            results2.append(np.mean(np.array(all_data[dataset][scan]['GraphMixer'][sampling][memory])))
        ax.errorbar(x=range(len(scans)), y=results, yerr=np.std(results), fmt='-', capsize=5, label="", color=colors[2*i])
        ax.errorbar(x=range(len(scans)), y=results2, yerr=np.std(results2), fmt='--', capsize=5, label="", color=colors[2*i])
        df_all[f"MR-{show_memorys[memory]}-atten-1l"] = results
        df_all[f"MR-{show_memorys[memory]}-atten-std-1l"] = np.std(results)
        df_all[f"MR-{show_memorys[memory]}-mixer-1l"] = results2
        df_all[f"MR-{show_memorys[memory]}-mixer-std-1l"] = np.std(results2)

        i += 1
    i = 0
    x_labels = scans
    ax.set_xticks(ticks=np.arange(len(x_labels)), labels=x_labels)
    ax.set_xlabel(f'# of Neighbors')
    ax.set_ylabel(f'MRR')

    ax = axes[1][k]
    scans = ['5x5', '5x10', '10x5', '10x10']
    if dataset == 'mooc' or dataset == 'uci':
        title_str = f'{dataset.upper()}'
    else:
        title_str = f'{dataset}'
    for memory in memorys:
        sampling = 're'
        means = []
        stds = []
        results = []
        results2 = []
        for scan in scans:
            results.append(np.mean(np.array(all_data[dataset][scan]['TGAT'][sampling][memory]) ))
            results2.append(np.mean(np.array(all_data[dataset][scan]['GraphMixer'][sampling][memory])))
        
        h1= ax.errorbar(x=range(len(scans)), y=results, yerr=np.std(results), fmt='-', capsize=5, 
                                   label=f"MR-{show_memorys[memory]}-atten", color=colors[2*i])
        h2=ax.errorbar(x=range(len(scans)), y=results2, yerr=np.std(results2), fmt='--', capsize=5, 
                                   label=f"MR-{show_memorys[memory]}-mixer", color=colors[2*i])
        df_all[f"MR-{show_memorys[memory]}-atten-2l"] = results
        df_all[f"MR-{show_memorys[memory]}-atten-std-2l"] = np.std(results)
        df_all[f"MR-{show_memorys[memory]}-mixer-2l"] = results2
        df_all[f"MR-{show_memorys[memory]}-mixer-std-2l"] = np.std(results2)
        if dataset == datasets[0]:
            handles.append(h1)
            handles.append(h2)
        i += 1
    i = 0
    df_all.to_csv(f'../tikz_data/mrr_aggr_{dataset}.csv', index=False)
    
    x_labels = scans
    ax.set_xticks(ticks=np.arange(len(x_labels)), labels=x_labels)
    ax.set_xlabel(f'# of Neighbors')
    ax.set_ylabel(f'MRR')
# plt.legend(loc='upper right', ncol=2)
# handles = [handle[1] for handle in handles]
fig.legend(handles=handles, frameon=False, bbox_to_anchor=(0.5, .01), loc='upper center',  ncol=3)
plt.tight_layout()
plt.savefig(f'../figures/mrr_aggr.pdf', bbox_inches='tight')
```
